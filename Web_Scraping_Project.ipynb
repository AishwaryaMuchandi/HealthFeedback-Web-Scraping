```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:
        product = 1
        result=[]
        j=0
        i=0
        while(i<len(nums)):
        # for i in range(len(nums)):
            if i == (len(nums)-1):
                j+=1
                product*=nums[i]
                result.append(product)
                product=1
                i=0
                continue
            elif i != j:
                product *=nums[i]
            i+=1
        return result
```
